#### Drop kolom
Kolom kode provinsi, nama provinsi, nama kabupaten kota, dan satuan didrop karena pada tabel keseluruhan fasilitas, semua berisikan data yang sama yang sudah distandarisasi. Namun, karena tetap memerlukan keterangan tersebut, hanya dua dari tiga tabel fasilitas yang didrop kolomnya.

```python
dropped_attr = ['kode_provinsi', 'nama_provinsi', 'nama_kabupaten_kota', 'satuan']
df_fasilitas_dump_truck = df_fasilitas_dump_truck.drop(dropped_attr,axis=1)
df_fasilitas_arm_truck = df_fasilitas_arm_truck.drop(dropped_attr,axis=1)

df_fasilitas_motor.info()
print("\n")
df_fasilitas_arm_truck.info()
```

#### Merge tabel
Semua tabel dilakukan merge untuk mendapatkan keseluruhan data fasilitas. Full outer digunakan agar mendapatkan seluruh data, yang ada di tabel. Diketahui bahwa record yang membedakan dari tiap tabel hanyalah jumlah fasilitasnya, sehingga tidak memerlukan praproses lebih lanjut.

```python
df_fasilitas = pd.merge(df_fasilitas_arm_truck, df_fasilitas_dump_truck, how="outer", on=["kode_kabupaten_kota", "tahun"])
df_fasilitas = pd.merge(df_fasilitas, df_fasilitas_motor, how="outer", on=["kode_kabupaten_kota", "tahun"])
df_fasilitas
```
